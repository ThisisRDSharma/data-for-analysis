```python
def custom_describe(df, column):
    r={}
    #creating a dictionary to store the values
    r['count'] = df[column].count()
    r['mean'] = df[column].mean()
    r['min'] = df[column].min()
    r['max'] = df[column].max()
    r['std'] = df[column].std()
    quantiles = df[column].quantile([0.25, 0.5, 0.75])
    r['25%'] = quantiles.loc[0.25]
    r['50%'] = quantiles.loc[0.5]
    r['75%'] = quantiles.loc[0.75]
    r['skewness']=df[column].skew()
    r['kurtosis']=df[column].kurt()
    print(type(r))

    return r
```

```python
result = custom_describe(df, 'id')
results = pd.DataFrame([result]).T
#T is to take transpose
results.columns = ['Custom Description']
print(results) 
```
